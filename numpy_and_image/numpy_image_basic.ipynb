# What is Image?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open("data/1.jpg")
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy() #coping the same image in the new variable
```

```python
pic_red.shape
```

```python
# R = 0, G = 1, B = 2
plt.imshow(pic_red[:,:,0])
```

```python
plt.imshow(pic_red[:,:,1]) #green Channel
```

```python
plt.imshow(pic_red[:,:,2]) #Blue channel
```

```python
# red channel values 0-255 full pure red
# value 0 means no red
plt.imshow(pic_red[:,:,0],cmap='gray')
```

```python
# Green Channel 0-255
# here 1 means no Green
plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
plt.imshow(pic_red[:,:,2],cmap='gray')
```

```python
pic_red[:,:,1]=0
plt.imshow(pic_red)
```

```python
pic_red[:,:,2]=0
plt.imshow(pic_red)
```
