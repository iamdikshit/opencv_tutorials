# Morphological Operator 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text = 'ABCD', org = (50,300), fontFace = font, fontScale = 5 , color = (255,255,255), thickness = 20)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

# Erosion 

```python
kernal = np.ones((5,5), dtype = np.uint8)
```

```python
kernal
```

```python
result = cv2.erode(img, kernal, iterations = 3)
```

```python
display_img(result)
```

```python
display_img(img)
```

```python
white_noise = np.random.randint(low = 0, high = 2, size = (600,600))
```

```python
display_img(white_noise)
```

```python
white_noise.max()
```

```python
white_noise.max()
```

```python
noise_image  = white_noise + img
```

```python
display_img(noise_image)
```

```python
opening = cv2.morphologyEx(noise_image, cv2.MORPH_OPEN, kernal)
```
