```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# brick image
brick = cv2.imread('data/9.jpg')
show_brick = cv2.cvtColor(brick, cv2.COLOR_BGR2RGB)

steve = cv2.imread('data/steve-jobs.jpg')
show_steve = cv2.cvtColor(steve, cv2.COLOR_BGR2RGB)

boy = cv2.imread('data/x4.jpg')
show_boy = cv2.cvtColor(boy, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_boy)
```

```python
hist_value = cv2.calcHist([show_boy], channels = [0], mask = None, histSize = [256], ranges = [0,256])
```

```python
hist_value.shape
```

```python
plt.plot(hist_value)
```

```python
hist_value = cv2.calcHist([show_brick], channels = [1], mask = None, histSize = [256], ranges = [0,256])
```

```python
plt.plot(hist_value)
```

```python
plt.imshow(show_brick)
```

```python
img = show_steve
```

```python
color = ('b','g','r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])
    plt.ylim([0,20000])

plt.title('HISTOGRAM FOR BRICK')
    
```
