# Contour Detection

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('data/s.jpg',0)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
ret, thresh = cv2.threshold(img, 127, 255, 0)
```

```python
plt.imshow(thresh, cmap = 'gray')
```

```python
thresh.shape
```

```python
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
hierarchy
```

```python
extern_contours = np.zeros(thresh.shape)
```

```python
extern_contours
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(extern_contours,contours,i,(255,0,0),-1)
```

```python
plt.imshow(extern_contours,cmap = 'gray')
```
