# Feature Matching (Brute force)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap = 'gray'):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
book = cv2.imread('data/book.jpg',0)
```

```python
 display(book)
```

```python
books = cv2.imread('data/books.jpg',0)
display(books)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(book,None)
kp2, des2 = orb.detectAndCompute(books, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)
```

```python
matches = bf.match(des1,des2)
```

```python
len(matches)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
```

```python
book_matches = cv2.drawMatches(book,kp1, books, kp2, matches[:],None,flags = 2)
```

```python
display(book_matches)
```
